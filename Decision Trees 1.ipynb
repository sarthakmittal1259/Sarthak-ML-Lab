```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
%matplotlib inline
```

```python
df=pd.read_csv('IRIS.csv')
```

```python
X=df.drop('species', axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state=42)
```

```python
#Decision Tree Classifier using the Gini index
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
clf_gini.fit(X_train, y_train)
y_pred_gini = clf_gini.predict(X_test)
print("Classification Report for Decision Tree using Gini Index:\n")
print(classification_report(y_test, y_pred_gini, target_names=df['species'].unique()))
```

```python
#Decision Tree Classifier using the Entropy
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf_entropy.fit(X_train, y_train)
y_pred_entropy = clf_entropy.predict(X_test)
print("Classification Report for Decision Tree using Entropy:\n")
print(classification_report(y_test, y_pred_entropy, target_names=df['species'].unique()))
```

```python
plt.figure(figsize=(20, 10))
```

```python
# Plot for Gini-based tree
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plot_tree(clf_gini, filled=True, feature_names=X.columns, class_names=df['species'].unique())
plt.title("Decision Tree using Gini Index")
plt.show()
```

```python
# Plot for Entropy-based tree
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 2)
plot_tree(clf_entropy, filled=True, feature_names=X.columns, class_names=df['species'].unique())
plt.title("Decision Tree using Entropy")
```
